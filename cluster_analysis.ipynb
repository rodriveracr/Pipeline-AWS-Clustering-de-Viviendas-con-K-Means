# Análisis de clusters - Viviendas

Este notebook carga `viviendas_local_with_clusters.csv` (generado por `local_run.py`) y muestra visualizaciones básicas: distribución de clusters y perfil promedio por cluster.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Cargar datos
df = pd.read_csv('viviendas_local_with_clusters.csv')
df.head()
```

```python
# Distribución de clusters
counts = df['cluster'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=counts.index, y=counts.values, palette='Set2')
plt.xlabel('Cluster')
plt.ylabel('Número de viviendas')
plt.title('Distribución de clusters')
plt.show()
```

```python
# Perfil promedio por cluster (features numéricas)
num_cols = ['Precio','Superficie','Antiguedad','Habitaciones','Precio_m2']
profile = df.groupby('cluster')[num_cols].mean().T
profile.plot(kind='bar', figsize=(10,6))
plt.title('Perfil promedio por cluster')
plt.ylabel('Valor medio')
plt.show()
```
